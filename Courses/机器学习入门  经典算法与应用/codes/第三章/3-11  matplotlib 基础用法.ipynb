# 绘制折线图

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

```python
x
```

```python
sin_y = np.sin(x)
```

```python
sin_y
```

```python
cos_y = np.cos(x)
```

```python
cos_y
```

## 同时绘制两条曲线
### 可以针对某一个具体函数指定图像颜色  color = "red" 
### 更改图像 的表现形式 linestyle="---"
### 更改坐标轴的范围 plt.xlim(-5,15)  plt.ylim(-3,10) 或者 plt.axis([-1,11,  -2,10]) 列表中前两个表示X范围，后两个是Y范围
### 声明坐标轴名称 plt.xlabel("X_label")   plt.ylabel("Y_label")   
### 图像曲线名称标注  label="sin(x)"  label="cos(x)"
### 图名称声明 plt.title("sin&&cos")

```python
plt.plot(x, sin_y, color="red",linestyle="--", label="sin(x)")
plt.plot(x, cos_y, color="yellow",label="cos(x)")
plt.xlabel("X_label")
plt.ylabel("Y_label")
plt.legend()
plt.title("sin&&cos show")
```

# 绘制散点图

## Scatter Plot

```python
plt.scatter(x,sin_y)
plt.scatter(x,cos_y)
```
